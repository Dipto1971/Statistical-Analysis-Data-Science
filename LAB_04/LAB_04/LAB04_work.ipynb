```python
import pandas as pd
!gdown --id 1g0Y27WxxsQCWWagNqUvajlre3tDOKsbM
import pandas as pd
df = pd.read_csv('syntactic_car_factory_data.csv')
# df = data frame
df
```

```python
from sklearn.preprocessing import LabelEncoder
df['Line'] = LabelEncoder().fit_transform(df['Line'])
df

df = pd.get_dummies(df, columns=['Sensor_Status', 'Shift'], drop_first=True)
df
```

```python
from sklearn.preprocessing import StandardScaler
df[['Sensor_1', 'Sensor_2', 'Sensor_3']] = StandardScaler().fit_transform(df[['Sensor_1', 'Sensor_2', 'Sensor_3']])
df
```

```python
import numpy as np
df['Sensor_8'] = np.log(df['Sensor_8'] + 1)
df
```
