```python
import pandas as pd
df = pd.read_csv('syntactic_car_factory_data.csv')
print(df.head())
```

```python
# grouped_df = df.groupby('Shift').mean()
df.groupby('Line').agg({'Sensor_1': 'mean', 'Sensor_2': 'sum'})
```

```python
pivot = pd.pivot_table(df, values= "Sensor_1",
                       index= "Shift",
                       columns= "Line",
                       aggfunc= "mean",
                       margins= True)
print(pivot)
# pivot function: powerful tool for summerizing data
```

```python
# Lambda Functions
df['Sensor_9'] = df['Sensor_9'].apply(lambda x: 1 if x>50 else 0)
df['Sensor_9']
```

```python
def adjust_sensor_value(x):
  if x>60:
    return x*1.2
  elif x<30:
    return x*0.8
  else:
    return x

df['Sensor_10'] = df['Sensor_10'].apply(adjust_sensor_value)
print(df.head())
```
