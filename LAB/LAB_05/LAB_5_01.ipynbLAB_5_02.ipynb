```python
import pandas as pd
df = pd.read_csv('syntactic_car_factory_data.csv')
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.plot(df.index, df['Sensor_1'], label= 'Sensor_1')
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Index')
plt.ylabel('Sensor_1 Values')
plt.title('Sensor_1 Values Over Time')
plt.legend()
plt.plot(df.index, df['Sensor_1'], label= 'Sensor_1')
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Sensor_2'], bins=40, color='skyblue', edgecolor='black')
plt.xlabel('Sensor_2 Values')
plt.ylabel('Frequency')
plt.title('Distribution of Sensor_2 Values')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Sensor_3'], df['Sensor_4'], alpha=0.7)
plt.xlabel('Sensor_3 Values')
plt.ylabel('Sensor_4 Values')
plt.title('Scatter Plot of Sensor_3 vs Sensor_4')
```

```python
# Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(x='Shift',y='Sensor_6', data=df)
plt.xlabel('Sensor_6 Values')
plt.ylabel('Frequency')
plt.title('Boxplot of Sensor_6 Values')
plt.show()

# Boxplot of a same columns' parameters
plt.figure(figsize=(10,6))
sns.boxplot(y='Sensor_6', data=df)
plt.xlabel('Sensor_6 Values')
plt.ylabel('Frequency')
plt.title('Boxplot of Sensor_6 Values')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sensor_Status', data=df)
plt.xlabel('Sensor Status')
plt.ylabel('Count')
plt.title('Count of Sensor Status')
plt.show()
```

```python
fig = px.line(df, x=df.index, y='Sensor_1', title='Interactive Line Plot of Sensor 1')
fig.show()
```
