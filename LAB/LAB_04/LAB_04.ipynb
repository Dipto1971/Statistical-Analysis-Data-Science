```python
import pandas as pd
df = pd.read_csv('syntactic_car_factory_data.csv')
# df = data frame
df.head()
```

```python
missing_values= df.isnull().sum()
print("Number of samples: ", len(df))
print(missing_values)
```

```python
missing_values = df.isnull().sum()
missing_values_percent = 100*df.isna().sum()/len(df)
missing_values_table = pd.concat([missing_values, missing_values_percent], axis=1)
missing_values_table
# Gives the percentage of missing values per column
```

```python
df['Sensor_1'] = df["Sensor_1"].fillna(df["Sensor_1"].mean())
df['Sensor_2'] = df["Sensor_2"].fillna(df["Sensor_2"].mean())
df['Sensor_3'] = df["Sensor_3"].fillna(df["Sensor_3"].mean())
df['Sensor_4'] = df["Sensor_4"].fillna(df["Sensor_4"].mean())
df['Sensor_5'] = df["Sensor_5"].fillna(df["Sensor_5"].mean())
df['Sensor_6'] = df["Sensor_6"].fillna(df["Sensor_6"].mean())
df['Sensor_7'] = df["Sensor_7"].fillna(df["Sensor_7"].mean())
df['Sensor_8'] = df["Sensor_8"].fillna(df["Sensor_8"].mean())
df['Sensor_9'] = df["Sensor_9"].fillna(df["Sensor_9"].mean())
df['Sensor_10'] = df["Sensor_10"].fillna(df["Sensor_10"].mean())

df.head(10)
```

```python
df['Shift'].fillna('Unknown', inplace=True)
df['Sensor_Status'].fillna(df['Sensor_Status'].mode()[0], inplace=True)
df
```

```python
# Machine Learning Framework: Scikit Learn
# Outlier Detection

q1 = df['Sensor_5'].quantile(0.25)
q3 = df['Sensor_5'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
print(lower_bound, upper_bound)

df = df[(df['Sensor_5'] > lower_bound) & (df['Sensor_5'] < upper_bound)]
df.head(10)

df_filter = df[(df['Sensor_5'] > lower_bound) & (df['Sensor_5'] < upper_bound)]
df_filter.head(10)

```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.05)
df['outlier'] = iso_forest.fit_predict(df[['Sensor_5']])
df = df[df['outlier'] != -1] #remove rows identified as outliers
df.head(1000)
```
